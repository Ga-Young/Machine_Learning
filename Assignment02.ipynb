<h3>Function Definition</h3>

$\quad f(x)=-x^3+2x^2+5$

<h3>Python Code (Taylor Approximation)</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# 1.Define  differentialble function that maps from real number to real number.
def f(x):             
    return -(x**3)+2*x**2+5
```

```python
# 2.Difine a domain of te function.
t=np.arange(-10,10,0.01) 

# 3.Plot the function

plt.title('Taylor Approximation')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()

plt.plot(t,f(t))
plt.show()
```

```python
# 4.Select a point within the domain.
p=3

# 5.Mark the selected point on the function.
plt.title('Taylor Approximation')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()

plt.plot(t,f(t))
plt.plot(p,f(p),'ro')

plt.show()
```

```python
# 6.Define the first-order Taylor approximation at the selected point.
x = [a for a in range(-10,10)]
f_hat_1=[f(p)+((-3)*p**2+4*p)*(x-p) for x in range(-10,10)]

plt.title('Taylor Approximation')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()

plt.plot(t,f(t))
plt.plot(p,f(p),'ro')

plt.plot(x,f_hat_1)
plt.show()
```

```python
# 7.Plot the Taylor approximation with the same domain of the original function.
f_hat=[-(x**3)+2*x**2+5 for x in range(-10,10)]

plt.title('Taylor Approximation')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid()

plt.plot(t,f(t))
plt.plot(p,f(p),'ro')

plt.plot(x,f_hat,'p--')
plt.show()
```
